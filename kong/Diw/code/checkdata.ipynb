```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df = pd.read_excel('C:\\Users\\kong0\\Desktop\\AI\\kong\\Diw\\data\\Gas1.xlsx')
data_X = df['message']
data_y = df['type']

#check data
type_counts = data_y.value_counts()
print(type_counts)
# Create the bar plot
plt.figure(figsize=(10, 6))
type_counts.plot(kind='bar')
plt.title('Number of Each Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
